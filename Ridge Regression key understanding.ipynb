How are coefficient affected?

```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=load_diabetes()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
df['Target']=data.target
```

```python
df.head()
```

```python
df.shape
```

```python
#Creating the indepednent and the dependent feature
x=data.data
y=data.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
coefs=[]
r2_scores=[]
for i in [0,10,100,1000]:
    reg=Ridge(alpha=i)
    reg.fit(x_train,y_train)
    coefs.append(reg.coef_.tolist())
    y_pred=reg.predict(x_test)
    r2_scores.append(r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(14,9))
plt.subplot(221)
plt.bar(data.feature_names,coefs[0])
plt.title('Alpha=0 ,r2_scores={}'.format(round(r2_scores[0],2)))
plt.subplot(222)
plt.bar(data.feature_names,coefs[1])
plt.title('Alpha=10,r2_scores={}'.format(round(r2_scores[1],2)))
plt.subplot(223)
plt.bar(data.feature_names,coefs[2])
plt.title('Alpha=100,r2_score={}'.format(round(r2_scores[2],2)))
plt.subplot(224)
plt.bar(data.feature_names,coefs[3])
plt.title('Alpha=1000,r2_score={}'.format(round(r2_scores[3],2)))
plt.show()
```

Higher coefficient are affected more

```python
alphas=[0,0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
coefs=[]
for i in alphas:
    reg=Ridge(alpha=i)
    reg.fit(x_train,y_train)
    coefs.append(reg.coef_.tolist())
```

```python
input_array=np.array(coefs)
```

```python
coef_df=pd.DataFrame(input_array,columns=data.feature_names)
coef_df['alpha']=alphas
coef_df.set_index('alpha')
```

```python
alphas=[0,0.0001,0.0005,0.001,0.005,0.1,0.5,1,5,10]
coefs=[]
for i in alphas:
    reg=Ridge(alpha=i)
    reg.fit(x_train,y_train)
    coefs.append(reg.coef_.tolist())
```

```python
input_array=np.array(coefs).T
```

```python
plt.figure(figsize=(15,8))
plt.plot(alphas,np.zeros(len(alphas)),color='black',linewidth=5)
for i in range(input_array.shape[0]):
    plt.plot(alphas,input_array[i],label=data.feature_names[i])
plt.legend()
```

Impact on bias and variance

```python
m=100
X=5*np.random.rand(m,1)-2
y=0.7*X**2-2*X+3+np.random.randn(m,1)
plt.scatter(X,y)
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X.reshape(100,1),y.reshape(100),test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=15)
X_train=poly.fit_transform(X_train)
X_test=poly.transform(X_test)
```

```python
from mlxtend.evaluate import bias_variance_decomp
alphas=np.linspace(0,30,100)
loss=[]
bias=[]
variance=[]
for i in alphas:
    reg=Ridge(alpha=i)
    avg_expected_loss,avg_bias,avg_var=bias_variance_decomp(
    reg,X_train,y_train,X_test,y_test,loss='mse',random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
plt.plot(alphas,loss,label='loss')
plt.plot(alphas,bias,label='Bias')
plt.plot(alphas,variance,label='variance')
plt.ylim(0,5)
plt.xlabel('Alpha')
plt.legend()
```

Effect of regularization on the loss function

```python
from sklearn.datasets import make_regression
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=30,random_state=42)
plt.scatter(X,y)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,y)
print(reg.coef_)
print(reg.intercept_)
```

```python
def call_loss(m,alpha):
    return (np.sum(y-m*X.ravel()+2.29)**2)+alpha*m*m
```

```python
def predict(m):
    return m*X-2.29
```

```python
m=np.linspace(-45,100,100)
plt.figure(figsize=(4,6))
for j in [0,10,20,30,40,50,100]:
    loss=[]
    for i in range(m.shape[0]):
        loss_i=call_loss(m[i],j)
        loss.append(loss_i)
    plt.plot(m,loss,label='alpha={}'.format(j))
plt.legend()
plt.xlabel("Alpha")
plt.ylabel("Loss")
plt.show()
```
